## GEN MIX RAW TABLES

### 22. GenMix 5-min (spp_physical.GenMix)

```python
df_genmix = sql_functions.download_df_from_sql_db(
    "SELECT IntervalEnd, GMTIntervalEnd, CoalTotal, WindTotal, SolarTotal, NaturalGasTotal, AverageActualLoad "
    "FROM spp_physical.GenMix "
    "WHERE DATE(IntervalEnd) = '2026-04-03' ORDER BY IntervalEnd")
print(df_genmix.shape)
# spot check: rows should be ~288 (5-min intervals * 24hr)
df_genmix.tail(10)
```

### 23. GenMixHourly (spp_physical.GenMixHourly)

```python
df_genmixhr = sql_functions.download_df_from_sql_db(
    "SELECT dt, hr, CoalTotal, WindTotal, SolarTotal, NaturalGasTotal, AverageActualLoad "
    "FROM spp_physical.GenMixHourly "
    "WHERE dt = '2026-04-01' ORDER BY hr")
print(df_genmixhr.shape)
# spot check: 24 rows expected
df_genmixhr
```

### 24. BAAZonalGenMix (spp_physical.BAAZonalGenMix) — E and W

```python
df_baa_genmix = sql_functions.download_df_from_sql_db(
    "SELECT IntervalEnd, BAA, CoalTotal, WindTotal, SolarTotal, NaturalGasTotal, AverageActualLoad\n",
    "FROM spp_physical.BAAZonalGenMix\n",
    "WHERE DATE(IntervalEnd) = '2026-04-01' ORDER BY IntervalEnd, BAA")
print(df_baa_genmix.shape)
# spot check: E + W rows; WindTotal_E + WindTotal_W should ≈ GenMix WindTotal
df_baa_genmix.head(10)
```

### 25. BAAZonalGenMixHourly (spp_physical.BAAZonalGenMixHourly) — E + W sum vs GenMixHourly

```python
df_baa_hr = sql_functions.download_df_from_sql_db(
    "SELECT dt, hr, BAA, WindTotal, SolarTotal, NaturalGasTotal, AverageActualLoad\n",
    "FROM spp_physical.BAAZonalGenMixHourly\n",
    "WHERE dt = '2026-04-01' ORDER BY hr, BAA")
print(df_baa_hr.shape)
# verify E+W sums match GenMixHourly
pivot = df_baa_hr.pivot_table(index=['dt','hr'], columns='BAA', values='WindTotal', aggfunc='sum')
pivot['total_EW'] = pivot.get('E', 0) + pivot.get('W', 0)
pivot = pivot.merge(df_genmixhr[['dt','hr','WindTotal']], on=['dt','hr'])
pivot['diff'] = pivot['total_EW'] - pivot['WindTotal']
print('Max E+W vs total diff (should be ~0):', pivot['diff'].abs().max())
pivot
```
